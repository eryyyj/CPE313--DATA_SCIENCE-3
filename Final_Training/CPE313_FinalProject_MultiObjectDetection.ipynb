```python
from pycocotools.coco import COCO
```

```python
transform = T.Compose([
    T.Resize(512),
    T.ToTensor(),
])

class BDD100KDataset(torch.utils.data.Dataset):
    def __init__(self, img_dir, ann_file, transforms=None):
        self.img_dir = img_dir
        self.coco = COCO(ann_file)
        self.ids = list(self.coco.imgs.keys())
        self.transforms = transforms

    def __getitem__(self, index):
        coco = self.coco
        img_id = self.ids[index]
        ann_ids = coco.getAnnIds(imgIds=img_id)
        anns = coco.loadAnns(ann_ids)

        # Load image
        path = coco.loadImgs(img_id)[0]['file_name']
        img = Image.open(os.path.join(self.img_dir, path)).convert('RGB')

        # Prepare the target
        num_objs = len(anns)

        boxes = []
        labels = []
        areas = []
        iscrowd = []

        for i in range(num_objs):
            xmin = anns[i]['bbox'][0]
            ymin = anns[i]['bbox'][1]
            width = anns[i]['bbox'][2]
            height = anns[i]['bbox'][3]
            boxes.append([xmin, ymin, xmin + width, ymin + height])
            labels.append(anns[i]['category_id'])  # assuming correct category ids
            areas.append(anns[i]['area'])
            iscrowd.append(anns[i].get('iscrowd', 0))

        boxes = torch.as_tensor(boxes, dtype=torch.float32)
        labels = torch.as_tensor(labels, dtype=torch.int64)
        areas = torch.as_tensor(areas, dtype=torch.float32)
        iscrowd = torch.as_tensor(iscrowd, dtype=torch.int64)

        target = {}
        target['boxes'] = boxes
        target['labels'] = labels
        target['image_id'] = torch.tensor([img_id])
        target['area'] = areas
        target['iscrowd'] = iscrowd

        if self.transforms is not None:
            img = self.transforms(img)

        return img, target

    def __len__(self):
        return len(self.ids)
```

```python
from torch.utils.data import DataLoader
```

```python
train_dir = 'bdd100kcoco/train'
train_ann = 'bdd100kcoco/annotations/_annotations_train.coco.json'

val_dir = 'bdd100kcoco/valid'
val_ann = 'bdd100kcoco/annotations/_annotations_val.coco.json'

test_dir = 'bdd100kcoco/test'
test_ann = 'bdd100kcoco/annotations/_annotations_test.coco.json'

train_ds = BDD100KDataset(train_dir, train_ann, transforms=transform)
val_ds = BDD100KDataset(val_dir, val_ann, transforms=transform)
test_ds = BDD100KDataset(test_dir, test_ann, transforms=transform)

train_dl = DataLoader(train_ds, batch_size=16, shuffle=True, num_workers=0, collate_fn=lambda x: tuple(zip(*x)))
val_dl = DataLoader(val_ds, batch_size=8, shuffle=True, num_workers=0, collate_fn=lambda x: tuple(zip(*x)))
test_dl = DataLoader(test_ds, batch_size=8, shuffle=True, num_workers=0, collate_fn=lambda x: tuple(zip(*x)))
```

```python
import os
```

```python
import torch.optim as optim

device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')
fasterrcnn_model.to(device)

params = [p for p in fasterrcnn_model.parameters() if p.requires_grad]
optimizer = optim.SGD(params, lr=0.0005, momentum=0.9)

num_epochs = 10
train_results = []
for epoch in range(num_epochs):
    print('training epoch:', epoch+1)
    fasterrcnn_model.train()
    print('training epoch:', epoch+1)
    for images, targets in train_dl:
        print('training')
        images = list(img.to(device) for img in images)
        targets = [{k: v.to(device) for k, v in t.items()} for t in targets]

        loss_dict = fasterrcnn_model(images, targets)
        losses = sum(loss for loss in loss_dict.values())
        
        optimizer.zero_grad()
        losses.backward()
        optimizer.step()

    train_results.append([epoch,losses.item()])
    print(f"Epoch #{epoch} Loss: {losses.item()}")
```

```python
for result in train_results:
    print(f'Epoch: {result[0]}, Loss: {result[1]}')
```

```python
with torch.no_grad():
    prediction = fasterrcnn_model([img])
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

img_cpu = img.cpu().permute(1,2,0).numpy()

fig, ax = plt.subplots(1, figsize=(12,9))
ax.imshow(img_cpu)

for box, label, score in zip(prediction[0]['boxes'], prediction[0]['labels'], prediction[0]['scores']):
    if score > 0.5:
        box = box.cpu().numpy() 
        xmin, ymin, xmax, ymax = box
        rect = patches.Rectangle((xmin, ymin), xmax - xmin, ymax - ymin,
                                 linewidth=2, edgecolor='red', facecolor='none')
        ax.add_patch(rect)
        ax.text(xmin, ymin, f'{label.item()} {score:.2f}', color='white',
                bbox=dict(facecolor='red', alpha=0.5))

plt.axis('off')
plt.show()
```
